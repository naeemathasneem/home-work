
1.Create a scatter plot of random points, where the x and y values are both generated randomly using NumPy's random module. Add labels to the x and y axes.

	
	


```python
import matplotlib.pyplot as plt
import numpy as np

#sample data 
x = np.random.normal(10,10,20)
y = np.random.normal(15,20,20)

#ploting a line graph
plt.scatter(x,y,label ="scatter plot",color = "purple",marker = "+")

#add labels and title
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("scatter plot")

#display the legend
plt.legend()

#display the plot
plt.show()
```

5.Create a histogram that shows the distribution of the ages of a group of people. Use the random module to generate some fake data for the ages.


```python
import matplotlib.pyplot as plt
import random

#sample data 
people = 100
ages=[random.randint(1,80) for i in range(people)]
#ploting a line graph
plt.hist(ages,bins = 25,edgecolor ="yellow",alpha =0.8,color = "purple",label = "Histogram") # alpha -> color thickeness

#add labels and title
plt.xlabel("ages")
plt.ylabel("frequency")
plt.title("ages of a group of people")

#display the legend
plt.legend()

#display the plot
plt.show() 

```

6.Create a stacked bar chart that shows the total number of cars sold by month, broken down by brand. Use the data provided below:

Month	Toyota	Honda	Ford

Jan	120	90	80

Feb	100	110	90

Mar	140	80	100

Apr	90	120	110

May	110	100	120


```python
import matplotlib.pyplot as plt

Month =  ["jan","feb","mar","apr","may"]
Toyota = [120,100,140,90,110]
Honda = [90,110,80,120,100]
Ford = [80,90,100,110,120]

fig =plt.figure()

ax = fig.add_axes([0,0,1,1]) 

b1 = ax.bar(Month,Toyota,label ="toyota",color ="yellow")
b2 = ax.bar(Month,Honda,label ="Honda",color ="black")
b3 = ax.bar(Month,Ford,label ="Ford",color ="green")
ax.legend(labels =("Toyota","Honda","Ford"),loc = "upper left")
ax.set_title("cars sold by month")
ax.set_xlabel("Month")
ax.set_ylabel("sales")
plt.show()
```

7.Create a scatter plot of the iris dataset, with the x-axis representing the sepal length and the y-axis representing the petal length. Color the points according to the species of iris.

	

```python
from sklearn import datasets

iris = datasets.load_iris()

import matplotlib.pyplot as plt

_, ax = plt.subplots()

scatter = ax.scatter(iris.data[:, 0], iris.data[:, 1], c=iris.target)

ax.set(xlabel=iris.feature_names[0], ylabel=iris.feature_names[1])

_ = ax.legend(scatter.legend_elements()[0], iris.target_names, loc="lower right", title="Classes")

#display the plot
plt.show()
```

9.. Create a stacked area chart that shows the percentage of time a person spends on different activities during the day. Use the data provided below:

Time of Day	Work	Exercise	Leisure

Morning	20%	5%	75%

Afternoon	60%	10%	30%

Evening	30%	5%	65%


```python
import matplotlib.pyplot as plt

time = ["morning","afternoon","evening"]

Work = [20,60,30]
Exercise=[5,10,5]
Leisure=[75,30,65]

plt.plot([],[],color ="g",label = "Work",linewidth = 5)
plt.plot([],[],color ="b",label = "Exercise",linewidth = 5)
plt.plot([],[],color ="r",label = "Leisure",linewidth = 5)

plt.stackplot(time,Work,Exercise,Leisure,colors =["g","b","r"])

plt.xlabel("time")
plt.ylabel("activity in percentage")
plt.title("percentage of time spends on different activities")
plt.legend()
plt.show()
```

10.Create a 3D scatter plot of random points, where the x, y, and z values are all generated randomly using NumPy's random module. Add labels to the x, y, and z axes.


```python
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d as pl3d
import numpy as np

#sample data 
x = np.random.normal(10,5,20)
y = np.random.normal(15,10,20)
z = np.random.normal(5,10,20)

#Creating figure
fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")
 
# Creating plot
ax.scatter3D(x, y, z, color = "blue")
plt.title("simple 3D scatter plot")
 

#display the plot
plt.show()

```
