```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size =0.3,random_state =0)
```

```python
# apply SVM model
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
svc = SVC()
svc.fit(x_train,y_train)
y_pred = svc.predict(x_test)
```

```python
print("Model accuracy :{0:0.3f}".format(accuracy_score(y_test,y_pred)))
```

```python
# confusion matrics
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
cm_matrix = pd.DataFrame(data = cm,columns =["Actual Positive:1","Actual Negative:0"],
                        index = ["Predict Positive :1","Predict Negeative:0"])
sns.heatmap(cm_matrix,annot=True,fmt ="d",cmap ="mako")
```

```python
# classifiction report 
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
# applying SVM model using sigmoid kernal functions
poly_svc = SVC(kernel ="sigmoid",C=1).fit(x_train,y_train)
y_pred =poly_svc.predict(x_test)
print("Model accuracy : {0:0.3f}".format(accuracy_score(y_test,y_pred)))

#regularization parameter (C) set to 1.it control the balance between maximizing the margin and minimizing the classification error

#K(x,y)=tanh((x^T)*y+r)
#x and y are the input feature vectors.
# is a parameter that controls the similarity of the features.r is another parameter that allows shifting the kernel.
# It can handle non-linear data transformations and decision boundaries.
```

```python
# applying SVM model using linear kernel function
# assumes that the relationship betwwen features is linear
Linear_Classifier =SVC(kernel="linear").fit(x_train,y_train)
y_pred =Linear_Classifier.predict(x_test)
print("Model accuracy with linear kernal : {0:0.3f}".format(accuracy_score(y_test,y_pred)))
```

```python
# applying SVM model using polynomial kernel function
poly_svc = SVC(kernel ="poly",C=1).fit(x_train,y_train)
y_pred =poly_svc.predict(x_test)
print("Model accuracy : {0:0.3f}".format(accuracy_score(y_test,y_pred)))

#K(x,y)=((x^T) *y+r) ^d
#x and y are the input feature vectors.
# is a parameter that controls the similarity of the features.
#r is another parameter that allows shifting the kernel.
#d is the degree of the polynomial.
```
